#**Merge**

```python
import pandas as pd
df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[2,4,6,8,10]})
df2 = pd.DataFrame({"A":[1,2,3,6,5],"C":[22,44,66,88,110]})
pd.merge(df1,df2,on="A")
```

```python
import pandas as pd
df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[2,4,6,8,10]})
df2 = pd.DataFrame({"A":[1,2,3,6,5],"C":[22,44,66,88,110]})
pd.merge(df1,df2,how = "inner")
```

```python
import pandas as pd
df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[2,4,6,8,10]})
df2 = pd.DataFrame({"A":[1,2,3,6,5],"C":[22,44,66,88,110]})
pd.merge(df1,df2,how = "left")
```

```python
import pandas as pd
df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[2,4,6,8,10]})
df2 = pd.DataFrame({"A":[1,2,3,6,5],"C":[22,44,66,88,110]})
pd.merge(df1,df2,how = "right")
```

```python
import pandas as pd
df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[2,4,6,8,10]})
df2 = pd.DataFrame({"A":[1,2,3,6,5],"C":[22,44,66,88,110]})
pd.merge(df1,df2,how = "outer")
```

```python
import pandas as pd
df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[2,4,6,8,10]})
df2 = pd.DataFrame({"A":[1,2,3,6,5],"C":[22,44,66,88,110]})
pd.merge(df1,df2,left_index=True,right_index=True)
```

#**join**

```python
import pandas as pd
df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[2,3,4,5,6]},index =[ "a","b","c","d","e"])
df2 = pd.DataFrame({"C":[10,20,30],"D":[20,30,40]},index =[ "a","b","c"])
df1.join(df2)
```

```python
import pandas as pd
df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[2,3,4,5,6]},index =[ "a","b","c","d","e"])
df2 = pd.DataFrame({"C":[10,20,30],"D":[20,30,40]},index =[ "a","b","c"])
df1.join(df2,how = "outer",lsuffix="_12",rsuffix="_23")
```
